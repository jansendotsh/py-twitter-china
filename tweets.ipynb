```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv("tweets.csv", sep=",")
df['tweet_time'] = pd.to_datetime(df['tweet_time'], format='%Y-%m-%d %H:%M')

dg = df.reset_index()['tweet_time'].value_counts().groupby(pd.Grouper(freq='M')).sum()
#dg.head(15)
dg.plot.line(x='tweet_time', xlabel="Tweet Date", ylabel="Number of Accounts", figsize=(15,10))
plt.show()
```
