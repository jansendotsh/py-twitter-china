```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv("dataset.csv", sep=",")

df.reset_index()['account_language'].value_counts().plot.bar(xlabel="Account Language", ylabel="Number of Accounts", figsize=(12,5))
plt.show()
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv("dataset.csv", sep=",")

df.reset_index()['account_language'].value_counts().plot.pie(ylabel="Percentage of Accounts by Language", figsize=(12,12))
plt.show()
```
