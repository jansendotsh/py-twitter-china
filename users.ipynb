```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv("users.csv", sep=",")

df.reset_index()['account_language'].value_counts().plot.bar(xlabel="Account Language", ylabel="Number of Accounts", figsize=(12,5))
plt.show()
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv("users.csv", sep=",")
df['account_creation_date'] = pd.to_datetime(df['account_creation_date'], format='%Y-%m-%d')

dg = df.reset_index()['account_creation_date'].value_counts().groupby(pd.Grouper(freq='M')).sum()
#dg.head(15)
dg.plot.line(x='account_creation_date', xlabel="Account Creation Date", ylabel="Number of Accounts", figsize=(15,10))
plt.show()
```
